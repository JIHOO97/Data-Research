```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

## Linear Graph

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x,y)

### add title
plt.title('Linear Graph')

# ### label the axis
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.plot(x,y)

plt.title('Linear Graph', fontdict={'fontsize': 25, 'fontname': 'Arial'})

plt.show()
```

```python
plt.plot(x,y)

plt.xticks([1,2,3])
plt.yticks([2,4,6])

plt.show()
```

```python
### add a color
plt.plot(x,y,'ro-.',markersize=5,linewidth=2,markeredgecolor='blue')

plt.show()
```

```python
x = [0,1,2,3,4]
y = [2,4,6,8,10]

plt.plot(x,y,'ro-',label='x')

x2 = np.arange(0,4.5,0.5)
y2 = x2**2
plt.plot(x2,y2,'bo--',label='x2')

plt.legend()

plt.show()
```

## Resizing a graph

```python
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x,y)

plt.show()
```

## Exercise 1

```python
gas_price_path = 'csvs/gasprices.csv'

gas_price_df = pd.read_csv(gas_price_path)
gas_price_df.head()
```

```python
gas_price_df.drop([0,1,2], inplace=True)
```

```python
gas_price_df.head()
```

```python
new_columns = []
for col in gas_price_df.iloc[0]:
    new_columns.append(col)

col_dict = dict(zip(gas_price_df.columns,new_columns))

gas_price_df = gas_price_df.rename(columns=col_dict)
gas_price_df = gas_price_df.drop([3])
gas_price_df = gas_price_df.reset_index(drop=True)
gas_price_df
```

```python
plt.figure(figsize=(5,3), dpi=100)

plt.title('Gas prices over time in USD', fontdict={'fontsize': 25, 'fontfamily': 'Arial'})
plt.plot(gas_price_df.Year, gas_price_df['South Korea'].astype('float32'),'bo-',label='South Korea')
plt.plot(gas_price_df.Year, gas_price_df['France'].astype('float32'),'ro-',label='France')
plt.plot(gas_price_df.Year, gas_price_df['Germany'].astype('float32'),'yo-',label='Germany')
plt.plot(gas_price_df.Year, gas_price_df['Italy'].astype('float32'),'go-',label='Italy')

plt.xticks(gas_price_df.Year[::3])

plt.xlabel('Year')
plt.ylabel('USD per gallon')

plt.legend()

plt.show()
```
