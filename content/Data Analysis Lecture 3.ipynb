```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gas_price_df = pd.read_csv('csvs/gasprices.csv')
gas_price_df.head()
```

```python
gas_price_df.drop([0,1,2], inplace=True)
```

```python
gas_price_df.head()
```

```python
keys = list(gas_price_df.columns)
values = list(gas_price_df.iloc[0])

col_dict = dict(zip(keys, values))

gas_price_df = gas_price_df.rename(columns=col_dict)
gas_price_df = gas_price_df.drop([3])
gas_price_df
```

```python
plt.title('Gas prices over time in UK')
plt.plot(gas_price_df.Year, gas_price_df.UK.astype('float32'))

plt.xticks(gas_price_df.Year[::3])

plt.xlabel('Year')
plt.ylabel('Gas price')

plt.show()
```

```python
fifa_df = pd.read_csv('csvs/players_20.csv')
fifa_df.head()
```

```python
left = fifa_df.loc[fifa_df['preferred_foot'] == 'Left'].count()[0]
right = fifa_df.loc[fifa_df['preferred_foot'] == 'Right'].count()[0]

plt.pie([left,right], labels=['Left','Right'], autopct='%.2f%%', explode=[.1,0])

plt.title('Foot preference of FIFA players')

plt.show()
```
