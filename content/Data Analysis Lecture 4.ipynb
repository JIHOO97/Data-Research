```python
import seaborn as sns
import pandas as pd

from datetime import datetime
```

```python
sales = pd.read_csv('csvs/sales.csv')
sales.head()
```

## What is the relationship between sales and profit?

```python
### relation plot
sns.relplot(x='Sales', y='Profit', data=sales)
```

```python
### hue parameter represents which column in the data frame, you want to use for color encoding
sns.relplot(x='Sales', y='Profit', data=sales, hue='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority', style='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority', style='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, size='Discount')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, size='Discount', sizes=(25,200))
```

## Line graph

```python
sns.relplot(x='Order Date', y='Sales', data=sales, kind='line')
```

```python
type(sales['Order Date'][0])
```

```python
sales['year'] = pd.to_datetime(sales['Order Date'])
sales.head()
```

```python
sales['year'] = pd.DatetimeIndex(sales['year']).year
sales = sales.sort_values(by='year')
sales.head()
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', ci='sd')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', estimator=None)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', estimator=sum)
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Product Category')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2)
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2, hue='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, col='Customer Segment', col_wrap=2, hue='Product Category')
```

## Boxplot

Mean, Median, Outliers

```python
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='box')
```

**Interquartile range: 25th to the 75th percentile**

```python
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='box', hue='Ship Mode')
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='box', hue='Ship Mode', hue_order=['Delivery Truck', 'Regular Air', 'Express Air'])
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data=sales, kind='box', col='Ship Mode')
```
